- phq4-1 (Câu hỏi PHQ4-1): Trong 2 tuần qua, bạn có thường xuyên cảm thấy lo lắng, căng thẳng hoặc bồn chồn không?
- phq4-2 (Câu hỏi PHQ4-2): Trong 2 tuần qua, bạn có thường xuyên cảm thấy không thể dừng hoặc kiểm soát việc lo lắng không?
- phq4-3 (Câu hỏi PHQ4-3): Trong 2 tuần qua, bạn có thường xuyên cảm thấy chán nản, tuyệt vọng hoặc buồn bã không?
- phq4-4 (Câu hỏi PHQ4-4): Trong 2 tuần qua, bạn có thường xuyên cảm thấy ít hứng thú hoặc niềm vui trong các hoạt động không?
- phq4_resp_mean (Thời gian trung bình trả lời PHQ4): Thời gian trung bình người dùng trả lời các câu hỏi PHQ4 (tính bằng giây).
- phq4_resp_median (Thời gian trung vị trả lời PHQ4): Thời gian trung vị người dùng trả lời các câu hỏi PHQ4 (tính bằng giây).
- phq4_score (Tổng điểm PHQ4): Tổng điểm PHQ4 (tổng của PHQ4-1 đến PHQ4-4).
- social_level (Mức độ giao tiếp xã hội): Trong ngày hôm nay, bạn dành nhiều thời gian ở một mình hay với người khác?
- sse3-1 (Thang đo lòng tự trọng - Câu 1): Hiện tại, tôi lo lắng về việc người khác nghĩ gì về tôi.
- sse3-2 (Thang đo lòng tự trọng - Câu 2): Hiện tại, tôi hài lòng với ngoại hình của mình.
- sse3-3 (Thang đo lòng tự trọng - Câu 3): Hiện tại, tôi cảm thấy mình thông minh như những người khác.
- sse3-4 (Thang đo lòng tự trọng - Câu 4): Nhìn chung, hiện tại tôi cảm thấy tốt về bản thân mình.
- sse3_resp_mean (Thời gian trung bình trả lời thang tự trọng): Thời gian trung bình trả lời các câu hỏi về lòng tự trọng (tính bằng giây).
- sse3_resp_median (Thời gian trung vị trả lời thang tự trọng): Thời gian trung vị trả lời các câu hỏi về lòng tự trọng (tính bằng giây).

```python
import pandas as pd
from datetime import datetime, timedelta

# Giả sử bạn đã có dataframe ge
# (với các cột: sleep_duration, sleep_start, sleep_end)
sensing = pd.read_csv("D:/UTH/kho dữ liệu và hệ thống hỗ trợ ra quyết định/project/data/main data/Sensing/sensing.csv")
def convert_to_time(epoch_value):
    """Chuyển đổi epoch (mỗi đơn vị = 7.5 phút kể từ 8PM) sang dạng HH:MM"""
    # Mốc bắt đầu: 8PM hôm trước
    base_time = datetime.strptime("20:00", "%H:%M")
    # Cộng thêm (epoch_value * 7.5 phút)
    new_time = base_time + timedelta(minutes=epoch_value * 7.5)
    # Nếu vượt qua 24h thì quay về mốc giờ trong ngày
    new_time = new_time - timedelta(days=new_time.day - base_time.day)
    return new_time.strftime("%H:%M")

# Tạo hai cột mới cho thời gian thực tế
sensing["sleep_start_time"] = sensing["sleep_start"].apply(convert_to_time)
sensing["sleep_end_time"] = sensing["sleep_end"].apply(convert_to_time)

# Kiểm tra kết quả
print(sensing[["sleep_start", "sleep_start_time", "sleep_end", "sleep_end_time", "sleep_duration"]].head())
```

```python
import pandas as pd

# === 1. Đọc file demographic (chứa uid và uid_new) ===
demo = pd.read_csv(
    "D:/UTH/kho dữ liệu và hệ thống hỗ trợ ra quyết định/project/data/main data/Demographics/demographics.csv",
    dtype={'uid': str, 'uid_new': str}
)

# Chuẩn hóa UID (bỏ khoảng trắng, chữ thường)
demo['uid'] = demo['uid'].str.strip().str.lower()

# Tạo dict map: uid cũ → uid_new
uid_map = demo.set_index('uid')['uid_new'].to_dict()

# === 2. Đọc các file dữ liệu chính (đảm bảo uid đọc dạng chuỗi) ===
ge = pd.read_csv(
    "D:/UTH/kho dữ liệu và hệ thống hỗ trợ ra quyết định/project/data/main data/EMA/general_ema.csv",
    dtype={'uid': str}
)
step = pd.read_csv(
    "D:/UTH/kho dữ liệu và hệ thống hỗ trợ ra quyết định/project/data/main data/Sensing/step.csv",
    dtype={'uid': str}
)
sensing = pd.read_csv(
    "D:/UTH/kho dữ liệu và hệ thống hỗ trợ ra quyết định/project/data/main data/Sensing/sensing.csv",
    dtype={'uid': str}
)

# Chuẩn hóa UID trong các file này
for df in [ge, step, sensing]:
    df['uid'] = df['uid'].astype(str).str.strip().str.lower()

# === 3. Map UID mới dựa vào dict ===
for name, df in [('general_ema', ge), ('step', step), ('sensing', sensing)]:
    df['uid_new'] = df['uid'].map(uid_map)
    print(f"{name}: {df['uid_new'].notna().sum()} mapped / {len(df)} total")

    # Nếu không map được thì giữ nguyên UID cũ
    df['uid'] = df['uid_new'].fillna(df['uid'])
    df.drop(columns=['uid_new'], inplace=True)

# === 4. Xuất lại file kết quả ===
ge.to_csv(
    "D:/UTH/kho dữ liệu và hệ thống hỗ trợ ra quyết định/project/data/main data/EMA/general_ema_mapped.csv",
    index=False
)
step.to_csv(
    "D:/UTH/kho dữ liệu và hệ thống hỗ trợ ra quyết định/project/data/main data/Sensing/step_mapped.csv",
    index=False
)
sensing.to_csv(
    "D:/UTH/kho dữ liệu và hệ thống hỗ trợ ra quyết định/project/data/main data/Sensing/sensing_mapped.csv",
    index=False
)

print("✅ Đã mapping UID mới (HSxxx) cho cả 3 file thành công và giữ nguyên UID cũ nếu không khớp!")
```

```python
import pandas as pd

# === 1. Đọc lại 3 file sạch ===
ge = pd.read_csv("D:/UTH/kho dữ liệu và hệ thống hỗ trợ ra quyết định/project/data/main data/EMA/general_ema.csv")
step = pd.read_csv("D:/UTH/kho dữ liệu và hệ thống hỗ trợ ra quyết định/project/data/main data/Sensing/step.csv")
sensing = pd.read_csv("D:/UTH/kho dữ liệu và hệ thống hỗ trợ ra quyết định/project/data/main data/Sensing/sensing.csv")

# === 2. Chuẩn hoá kiểu dữ liệu cột day (nếu cần) ===
for df in [ge, step, sensing]:
    if 'day' in df.columns:
        df['day'] = pd.to_datetime(df['day'], errors='coerce').dt.date

# === 3. Giữ lại các cột cần thiết ===
ge = ge[['uid', 'day']].dropna(subset=['uid', 'day']).drop_duplicates()
step = step[['uid', 'day']].dropna(subset=['uid', 'day']).drop_duplicates()
sensing = sensing[['uid', 'day']].dropna(subset=['uid', 'day']).drop_duplicates()

# === 4. Tìm giao nhau giữa ba bảng theo cả uid và day ===
common_uid_day = ge.merge(step, on=['uid', 'day'], how='inner').merge(sensing, on=['uid', 'day'], how='inner')

# === 5. In kết quả tổng hợp ===
print("🟢 Tổng số dòng (uid, day) trong general_ema:", len(ge))
print("🟢 Tổng số dòng (uid, day) trong step:", len(step))
print("🟢 Tổng số dòng (uid, day) trong sensing:", len(sensing))
print("-------------------------------------------------")
print("✅ Số dòng (uid, day) xuất hiện trong cả 3 file:", len(common_uid_day))



```
